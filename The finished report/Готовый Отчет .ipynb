### Для создания таблицы `Готовый отчет` все готово. Соединю файлы

В отчете не нужны некоторые колонки, удалю их. Также, есть дубли столбцов, их тоже удалю

```python
ГотовыйОтчет = ГотовыйОтчет.drop(columns =['salesvalue', 'purchasevalue', 'week', 'chain_technical_y', 'chain_technical_x', 'Unnamed: 8' ])
ГотовыйОтчет
```

Посмотрю общую инфо по столбцам

```python
ГотовыйОтчет.info()
```

Необходимо сформировать столбцец `salesvolume` и сдлеать вычисления из двух столбцов, у которых неверный тип - строчной. Исправлю это

```python
ГотовыйОтчет['weight'] = pd.to_numeric(ГотовыйОтчет['weight'], errors='coerce' )
ГотовыйОтчет['weight'].info()
```

```python
ГотовыйОтчет['weight'].isna().sum()
```

```python
ГотовыйОтчет['weight'] = ГотовыйОтчет['weight'].astype(int)
```

```python
ГотовыйОтчет['salesitem'] = pd.to_numeric(ГотовыйОтчет['salesitem'], errors='coerce' )
ГотовыйОтчет['salesitem'].info()
```

```python
display(ГотовыйОтчет['salesitem'].isna().sum())
ГотовыйОтчет['salesitem'] = ГотовыйОтчет['salesitem'].fillna(0)
ГотовыйОтчет['salesitem'].isna().sum()
```

```python
ГотовыйОтчет['salesitem'] = ГотовыйОтчет['salesitem'].astype(int)
```

Типы исправила на числовые, теперь сформирую новый столбце

```python
ГотовыйОтчет['salesvolume'] = ГотовыйОтчет['salesitem']*ГотовыйОтчет['weight']/1000
```

Для дальнейшей проверки отчета и выводов, нужно исправить тип еще в некоторых столбцах

```python
ГотовыйОтчет['salesvalueVAT'] = [x.replace(',', '.') for x in ГотовыйОтчет['salesvalueVAT']]

ГотовыйОтчет['salesvalueVAT'] =  ГотовыйОтчет['salesvalueVAT'].astype(float)

ГотовыйОтчет['salesvalueVAT'].info()
```

```python
ГотовыйОтчет['purchasevalueVAT'] = [x.replace(',', '.') for x in ГотовыйОтчет['purchasevalueVAT']]

ГотовыйОтчет['purchasevalueVAT'] =  ГотовыйОтчет['purchasevalueVAT'].astype(float)

ГотовыйОтчет['purchasevalueVAT'].info()
```
